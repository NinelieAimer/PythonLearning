```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import tushare as ts
```

```python
#读取文件，然后设置题头，这个非常重要
unames=['user_id','gender','age','occupation','zip']
users=pd.read_table(r'D:\learning\PythonLearning\analysisData\Review&Test\pydata-book\datasets\movielens\users.dat',
                    sep='::',header=None,names=unames)
```

```python
users.head(n=10)
```

```python
#读取rating文件
rnames=['user_id','movie_id','rating','timestamp']
ratting=pd.read_table(r'D:\learning\PythonLearning\analysisData\Review&Test\pydata-book\datasets\movielens\ratings.dat',
                      header=None,names=rnames,sep='::')
```

```python
ratting.head(n=100)
```

```python
mnames=['movie_id','title','genres']
movies=pd.read_table(r'D:\learning\PythonLearning\analysisData\Review&Test\pydata-book\datasets\movielens\movies.dat',
                     sep='::',header=None,names=mnames)
```

```python
movies.head()
```

```python
data=pd.merge(pd.merge(left=ratting,right=users,how='outer',on='user_id'),movies,on='movie_id',how='outer')
```

```python
data.head()
```

```python
data.sort_values('user_id')
```

```python
data.iloc[0]
```

```python
mean_ratings=data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
```

```python
mean_ratings.head()
```

```python
data.describe()
```

```python
ratting_by_title=data.groupby(['title']).size()
```

```python
ratting_by_title
```

```python
mean_ratings=mean_ratings[ratting_by_title>=250]
```

```python
top_female_rating=mean_ratings.sort_values(by='F',ascending=False)
```
