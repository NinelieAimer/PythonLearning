```python
import tushare as ts
import numpy as np
import pandas as pd
```

```python
pro=ts.pro_api('9526a2198c86fb0a72fa552f74ba1140033871f755dcd769e3134b7b')
```

```python
df=pro.daily(ts_code='000550.SZ',start_date='20110101',end_date='20190702')
```

```python
df.head()
```

```python
new=pd.to_datetime(df['trade_date'])
```

```python
df['trade_date']=new
```

```python
df.head()
```

```python
df.set_index(['trade_date'],inplace=True)
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
df.resample('M').agg({'close':'mean'})
```

```python
df.tail()
```

```python
df['Repay']=0.0
for i in range(1,len(df)):
    df['Repay'][i]=np.log(df['close'][i]/df['close'][i-1])
```
