```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df=pd.read_excel('./hprice1.xlsx')
```

```python
x=df[['sqrft','bdrms']]
y=df['price']
x=sm.add_constant(x)
```

```python
model=sm.OLS(y,x)
result=model.fit()
```

```python
result.summary()
```

```python
confidence_interval=result.conf_int()   #confidence interval
SSE=result.ess  #SSE
SSR=result.ssr  #SSR
SST=result.centered_tss #SST
BETA=result.params  #BETA
var=result.mse_resid    #varience
```

```python
result.tvalues  #compute
```

```python
result.t_test('bdrms=0')
```
