```python
import numpy as np 
import pandas as pd 
import torch
```

```python
#read the file
df=pd.read_excel('hprice1.xlsx')
df=df[['price','sqrft','bdrms']]
```

```python
#initialize the inputs and change dataframe to tensor
inputs=df[['sqrft','bdrms']].values
inputs=torch.from_numpy(inputs)
```

```python
target=df['price'].values
target=torch.from_numpy(target)
```

```python
#initialize beta and alpha
beta=torch.rand(2,requires_grad=True)
alpha=torch.rand(len(inputs),requires_grad=True)
```

```python
#change the dtype of inputs and target, it's really important
inputs=inputs.float()
target=target.float()
```

```python
#define the target function
def model(x):
    return alpha+x@beta
```

```python
#define the loss function
def mse(ori,pre):
    diff=ori-pre
    return torch.sum(diff*diff)/diff.numel()
```

```python
#computing the result
for i in range(10000):

    #compute the loss and figure out the derivatives
    pred=model(inputs)
    loss=mse(target,pred)
    loss.backward()

    #attention that you should use the torch.no_grad() because you don't want to track the compute of the beta and alpha
    with torch.no_grad():
        beta-=beta.grad* (10**(-10))
        alpha-=alpha.grad*(10**(-10))

        #initialize the beta and alpha grad
        beta.grad.zero_()
        alpha.grad.zero_()
```
